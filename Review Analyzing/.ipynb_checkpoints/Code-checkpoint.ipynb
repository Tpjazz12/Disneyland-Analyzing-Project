# The Vader Model

### Vader tokenizes each word in a given text and marks is as positive, negative, or neutral. It then weights this to calculate a compound which determines if the general sentiment is positive or negatve. 

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
```

```python
#reviews = df2['reviews']
df2['scores'] = df2['reviews'].apply(lambda reviews: sia.polarity_scores(str(reviews)))
```

```python
df2
```

```python
df2['compound']  = df2['scores'].apply(lambda score_dict: score_dict['compound'])

df2.head()
```

```python
#Here I set the limits so that each review is categorized by their star rating. 1&2 being negative, 3 being neuttral and 4&5 being positive.
def condition(x):
    if x>=.2:
        return 'pos'
    elif x<=(-.2):
        return 'neg'
    else:
        return 'neu'
```

```python
df2['comp_score'] = df2['compound'].apply(condition)
df2
```

```python
#Replace ratings with something we can compare to vader
df2['Given Sentiment Rating'] = df.loc[:, 'Rating']
df2['Given Sentiment Rating'] = df2['Given Sentiment Rating'].astype(str)
df2['Given Sentiment Rating'] = df2['Given Sentiment Rating'].replace(['5'],'pos')
df2['Given Sentiment Rating'] = df2['Given Sentiment Rating'].replace(['4'],'pos')
df2['Given Sentiment Rating'] = df2['Given Sentiment Rating'].replace(['3'],'neu')
df2['Given Sentiment Rating'] = df2['Given Sentiment Rating'].replace(['2'],'neg')
df2['Given Sentiment Rating'] = df2['Given Sentiment Rating'].replace(['1'],'neg')
```

```python
#Comparing the vader results with the star rating so that I can compute accuracy. 
df2['vader accuracy'] = df2['Given Sentiment Rating'] == df2['comp_score']
df2
```

# Roberta Model

```python
from transformers import AutoTokenizer
from transformers import AutoModelForSequenceClassification
from scipy.special import softmax
```

```python
df3 = df2.assign(row_number=range(len(df)))
df3['reviews'] = df3['reviews'].astype(str)
df4 = df3.drop(df3.index[6285])
```

```python
MODEL = f"cardiffnlp/twitter-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = AutoModelForSequenceClassification.from_pretrained(MODEL)
```

```python
def polarity_scores_roberta(example):
  encoded_text = tokenizer(example, return_tensors='pt')
  output = model(**encoded_text)
  scores = output[0][0].detach().numpy()
  scores = softmax(scores)
  scores_dict = {
      'roberta_neg' : scores[0],
      'roberta_neu' : scores[1],
      'roberta_pos' : scores[2]
  }
  return scores_dict
```

```python
res = {}
for i, row in tqdm(df4.iterrows(), total = len(df4)):
  try:
      text = row['reviews']
      myid = row['row_number']
      vader_result = sia.polarity_scores(text)
      vader_result_rename ={}
      for key, value in vader_result.items():
         vader_result_rename[f"vader_{key}"] = value
      roberta_result = polarity_scores_roberta(text)
      both = {**vader_result_rename,**roberta_result}
      res[myid] = both
  except RuntimeError:
     print(f'Broke for row {myid}')
```

```python
# here I create the dataframe for our results
results_df = pd.DataFrame(res).T
results_df.columns
results_df = results_df.reset_index().rename(columns={'index': 'row_number'})
results_df = results_df.merge(df3, how= 'left')
```

```python
results_df
```

```python
final1 = results_df.drop(['row_number', 'vader_compound', 'Rating', 'asin', 'reviews', 'scores'], axis = 1)
final = final1.rename(columns={"compound": "vader numerical compound", "comp_score": "vader sentiment", "Accuracy":"vader accuracy"})

#sub dataframes just in case
vader = final.drop(['roberta_neu', 'roberta_pos', 'roberta_neg'], axis = 1)
roberta = final.drop(['vader_neu', 'vader_pos', 'vader_neg'], axis = 1)
```

## ROBERTA does not calculate a compund. The ROBERTA determines the sentiment by comparing the 3 probabilities [pos, neg, neu] and the highest becomes the label used to mark the review. 

```python
#Create the ROBERTA result column 
final['ROBERTA result'] = results_df[['roberta_pos', 'roberta_neg', 'roberta_neu']].idxmax(axis=1)
final['ROBERTA result'] = final['ROBERTA result'].replace(['roberta_pos'],'pos')
final['ROBERTA result'] = final['ROBERTA result'].replace(['roberta_neu'],'neu')
final['ROBERTA result'] = final['ROBERTA result'].replace(['roberta_neg'],'neg')
```

```python
#Accuracy of ROBERTA
final['ROBERTA accuracy'] = final['Given Sentiment Rating'] == final['ROBERTA result']
extracted_col = df2["Rating"]
final.insert(1, "Rating", extracted_col)
final2 = final.loc[:,['vader_pos','vader_neu','vader_neg', 'vader numerical compound', 'roberta_pos', 'roberta_neu', 'roberta_neg','vader sentiment', 'ROBERTA result', 'Rating', 'Given Sentiment Rating', 'vader accuracy', 'ROBERTA accuracy']]
final2
```

## changing categorical values to numerical

```python
#ROBERTA
final2['ROBERTA result'] = final2['ROBERTA result'].replace(['pos'],'1')
final2['ROBERTA result'] = final2['ROBERTA result'].replace(['neu'],'0')
final2['ROBERTA result'] = final2['ROBERTA result'].replace(['neg'],'-1')

final2['ROBERTA accuracy'] = final2['ROBERTA accuracy'].replace([True],'1')
final2['ROBERTA accuracy'] = final2['ROBERTA accuracy'].replace([False],'0')

#VADER
final2['vader sentiment'] = final2['vader sentiment'].replace(['pos'],'1')
final2['vader sentiment'] = final2['vader sentiment'].replace(['neu'],'0')
final2['vader sentiment'] = final2['vader sentiment'].replace(['neg'],'-1')

final2['vader accuracy'] = final2['vader accuracy'].replace([True],'1')
final2['vader accuracy'] = final2['vader accuracy'].replace([False],'0')


#Results and given
final2['Given Sentiment Rating'] = final2['Given Sentiment Rating'].replace(['pos'],'1')
final2['Given Sentiment Rating'] = final2['Given Sentiment Rating'].replace(['neu'],'0')
final2['Given Sentiment Rating'] = final2['Given Sentiment Rating'].replace(['neg'],'-1')

#rating
final2['Rating'] = final2['Rating'].replace(['5'],'5 star')

final2['Rating'] = final2['Rating'].replace(['4'],'4 star')

final2['Rating'] = final2['Rating'].replace(['3'],'3 star')

final2['Rating'] = final2['Rating'].replace(['2'],'2 star')

final2['Rating'] = final2['Rating'].replace(['1'],'1 star')
```
